```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO
from PyPDF2 import PdfFileMerger, PdfFileReader
import re
import pandas as pd
import xlrd
import csv
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
class PdfConverter:
    def __init__(self, file_path):
            self.file_path = file_path

    # convert pdf file to a string which has space among words
    def convert_pdf_to_txt(self):
        rsrcmgr = PDFResourceManager()
        retstr = StringIO()
        codec = 'utf-8'
        laparams = LAParams()
        device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
        fp = open(self.file_path, 'rb')
        interpreter = PDFPageInterpreter(rsrcmgr, device)
        password = ""
        maxpages = 0
        caching = True
        pagenos = set()
        for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password, caching=caching,
                                      check_extractable=True):
            interpreter.process_page(page)
        fp.close()
        device.close()
        str = retstr.getvalue()
        retstr.close()
        return str

    # convert pdf file text to string and save as a text_pdf.txt file

    def save_convert_pdf_to_txt(self):
        content = self.convert_pdf_to_txt()
        txt_pdf = open(r'D:\StudyMaterials\Sem2\DataScience\Assignment1\Final_report.txt', 'wb')
        txt_pdf.write(content.encode('utf-8'))
        txt_pdf.close()
  
def merge_pdf_file():
    merger = PdfFileMerger()
    merger.append(PdfFileReader(open(r'D:\StudyMaterials\Sem2\DataScience\Assignment1\Beyond_Fintech_-_A_Pragmatic_Assessment_of_Disruptive_Potential_in_Financial_Services.pdf', 'rb')))
    merger.append(PdfFileReader(open(r'D:\StudyMaterials\Sem2\DataScience\Assignment1\WEF_A_Blueprint_for_Digital_Identity.pdf', 'rb')))
    merger.append(PdfFileReader(open(r'D:\StudyMaterials\Sem2\DataScience\Assignment1\WEF_The_future__of_financial_services.pdf', 'rb')))
    merger.append(PdfFileReader(open(r'D:\StudyMaterials\Sem2\DataScience\Assignment1\WEF_The_future_of_financial_infrastructure.pdf', 'rb')))
    merger.write(r'D:\StudyMaterials\Sem2\DataScience\Assignment1\A.pdf')

#Count word and save to excel    
def word_list_excel():
    data = open(r'D:\StudyMaterials\Sem2\DataScience\Assignment1\Final_report.txt', 'r', encoding='utf-8').read()
    data = data.replace('\n', ' ')
    data = data.lower()
    data = re.sub('\.\.+', '', data)
    data = re.sub('\…\…+', '', data)
    remove_special_chars = data.translate({ord(c): '' for c in "\.{2,}!@#$%^&*()[]\\{};:,./<>?\|`~-=_+\"\“\”"})
    remove_special_chars = re.sub(r'[^a-zA-Z\s]+', '', remove_special_chars)
    stop_words = set(stopwords.words('english'))
    filtered_sentence = []
    word_tokens = word_tokenize(remove_special_chars)

    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)
    
    word_list = filtered_sentence
    wordfreq = {}

    for a in word_list:
        if a not in wordfreq:
            wordfreq[a] = 0
        wordfreq[a] += 1

    df = pd.DataFrame.from_dict(wordfreq,columns=['keyWords'], orient='index')
    #print(df)
    #writer = pd.ExcelWriter ('e.xlsx', engine='openpyxl')
    writer = pd.ExcelWriter(r'D:\StudyMaterials\Sem2\DataScience\Assignment1\word_count.xlsx')
    df.to_excel(writer, 'Sheet1')
    writer.save()

def csv_from_excel():
    wb = xlrd.open_workbook(r'D:\StudyMaterials\Sem2\DataScience\Assignment1\word_count.xlsx')
    sh = wb.sheet_by_name('Sheet1')
    your_csv_file = open('word_count.csv', 'w')
    wr =csv.writer(your_csv_file, quoting= csv.QUOTE_ALL)

    for rownum in range(sh.nrows):
        wr.writerow(sh.row_values(rownum))

    your_csv_file.close()
    print("csv printed")
    
if __name__ == '__main__':

# PART 1
# Merge PDF File
    ##merge_pdf_file()

    # Convert Pdf to TextFile
    #pdfConverter = PdfConverter(file_path=r'D:\StudyMaterials\Sem2\DataScience\Assignment1\A.pdf')
    #pdfConverter.save_convert_pdf_to_txt()

    # Saving Word count to excel sheet
    word_list_excel()
    csv_from_excel()
```
